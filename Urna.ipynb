```python
from os import system
from time import sleep

continua = 'S'
cont12 = 0
cont15 = 0
cont18 = 0
cont20 = 0
cont22 = 0
contnulo = 0

while continua.lower() == 's':
    system('clear')
    print('=' * 57 + f'{"Eleições Alvaristas":^57}' + '=' * 57)
    print('\nCandidatos:\n\nAguinaldo Jr. - 12\nNivaldex - 15\nViviane - 18\nRafael Rosetti - 20\nMilton Goya - 22\nNulo - 0')

    precod = int(input('\nVOTO PARA PRESIDENTE: '))

    if precod == 12:
        system('clear')
        print(f'''NOME: Aguinaldo Jr.
VICE: Paulo
PARTIDO: PY''')

        confirma = input('\nConfirmar o voto? [S/N] ')
        if confirma.lower() == 's':
            cont12 += 1

        continua = input('\nDeseja continuar? (S/N): ')
        if continua.lower() != 's':
            continua = 'N'
    elif precod == 15:
        system('clear')
        print(f'''NOME: Nivaldex
VICE: Vivi
PARTIDO: ARD''')

        confirma = input('\nConfirmar o voto? [S/N] ')
        if confirma.lower() == 's':
            cont15 += 1

        continua = input('\nDeseja continuar? (S/N): ')
        if continua.lower() != 's':
            continua = 'N'
    elif precod == 18:
        system('clear')
        print(f'''NOME: Viviane
VICE: Fernando
PARTIDO: NEGS''')

        confirma = input('\nConfirmar o voto? [S/N] ')
        if confirma.lower() == 's':
            cont18 += 1

        continua = input('\nDeseja continuar? (S/N): ')
        if continua.lower() != 's':
            continua = 'N'
    elif precod == 20:
        system('clear')
        print(f'''NOME: Rafael Rosetti
VICE: Marcão
PARTIDO: IA''')

        confirma = input('\nConfirmar o voto? [S/N] ')
        if confirma.lower() == 's':
            cont20 += 1

        continua = input('\nDeseja continuar? (S/N): ')
        if continua.lower() != 's':
            continua = 'N'
    elif precod == 22:
        system('clear')
        print(f'''NOME: Milton Goya
VICE: Neusa
PARTIDO: BDD''')

        confirma = input('\nConfirmar o voto? [S/N] ')
        if confirma.lower() == 's':
            cont22 += 1

        continua = input('\nDeseja continuar? (S/N): ')
        if continua.lower() != 's':
            continua = 'N'
    elif precod == 0:
        system('clear')
        print('VOTO NULO')

        confirma = input('\nConfirmar o voto nulo? [S/N] ')
        if confirma.lower() == 's':
            contnulo += 1

        continua = input('\nDeseja continuar? (S/N): ')
        if continua.lower() != 's':
            continua = 'N'
    else:
        print('\nNúmero inválido! Tente novamente.')
        sleep(2)

system('clear')
print('=' * 57 + f'{"Resultado":^57}' + '=' * 57)
print(f'Aguinaldo Jr.: {cont12} votos')
print(f'Nivaldex: {cont15} votos')
print(f'Viviane: {cont18} votos')
print(f'Rafael Rosetti: {cont20} votos')
print(f'Milton Goya: {cont22} votos')
print(f'Votos Nulos: {contnulo} votos')

sleep(10)
```
